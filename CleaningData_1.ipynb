### 9. Remove Duplicate

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

### 10. Remove Id column

```python
df.drop(columns=['id'], inplace=True)
df.head()
```

### 11. Age days into year

```python
df['age'] = (df['age'] / 365).astype(int)
df.tail()
```

### 12. Fix Dtatype

```python
cat_cols = ['gender','cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio']
for col in cat_cols:
    df[col] = df[col].astype('category')
df.info()
```

### 13-14. Remove extreme outliers in height & weight

```python
df = df[(df['height'] >= 120) & (df['height'] <= 220)]
```

```python
df = df[(df['weight'] >= 30) & (df['weight'] <= 200)]
```

### 15. Filter out impossible & outliers blood pressure values 

```python
df = df[df['ap_hi'] >= df['ap_lo']]
df = df[(df['ap_hi'] > 0) & (df['ap_lo'] > 0)]
df = df[(df['ap_hi'] < 300) & (df['ap_lo'] < 200)]
```

### 16. Create BMI column

```python
df['bmi'] = df['weight'] / ((df['height']/100)**2)
df.head()
```

### 17. Unique Values

```python
print(df['gender'].unique())
print(df['cholesterol'].unique())
print(df['gluc'].unique())
print(df['smoke'].unique())
print(df['alco'].unique())
print(df['active'].unique())
print(df['cardio'].unique())
```

### 18. Remove height weight or bmi (Redudandunt)

```python
df = df.drop(['height', 'weight'], axis=1)
df.head()
```

### 19. Normalize or scale numerical columns

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_cols = ['age','ap_hi','ap_lo','bmi']
df[num_cols] = scaler.fit_transform(df[num_cols])
df.head()
```

### 20. Checking for imbalaced

```python
df['cardio'].value_counts(normalize=True)
```

### 21. save file

```python
df.to_csv('cleaned_cardio.csv', index=False)
```
